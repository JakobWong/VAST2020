```python
import os

root_dir = 'MC2-Image-Data'

res_list = {}

for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith('.txt'):
            
            # Open a file: file
            txt_file = open(os.path.join(root,file),'r',encoding='cp1252')

            # read all lines at once
            all_of_it = txt_file.readline()
            
            # close the file
            txt_file.close()
            
            filecode = file.replace('caption.txt','')
            res_list[filecode] = all_of_it
#             res_list.append(os.path.join(root,file))
            
```
